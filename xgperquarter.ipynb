```python
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt 
from mplsoccer import  Sbopen
```

```python
parser = Sbopen()
df_match = parser.match(competition_id=43, season_id=3)
team = "Mexico"

match_id=df_match[df_match["home_team_name"] == team].iloc[0]['match_id']
```

```python
match_id
```

```python
df = parser.event(match_id)[0]
df.columns
```

```python
df[['minute','second','player_id','player_name','type_id','type_name','shot_statsbomb_xg','outcome_name','team_name','team_id']]
```

```python
shot = df[df['type_name'] == 'Shot']
shot[['minute','second','player_id','player_name','type_id','type_name','shot_statsbomb_xg','outcome_name','team_name','team_id']]
```

```python
a_xG = [0]
h_xG = [0]
a_min= [0]
h_min= [0]

hteam = shot['team_name'].iloc[0]
ateam = shot['team_name'].iloc[-1]


for index,row in shot.iterrows():
    if row['team_name'] == ateam:
        a_xG.append(row['shot_statsbomb_xg'])
        a_min.append(row['minute'])
    if row['team_name'] == hteam:
        h_xG.append(row['shot_statsbomb_xg'])
        h_min.append(row['minute'])

def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

a_cumulative = nums_cumulative_sum(a_xG)
h_cumulative = nums_cumulative_sum(h_xG)

alast = round(a_cumulative[-1],2)
hlast = round(h_cumulative[-1],2)
```

```python
print(ateam)
print (h_min)
print (h_cumulative)
```

```python
fig,ax= plt.subplots(figsize= (10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'

ax.grid(ls='dotted', lw=.5, color='lightgrey',axis='y',zorder=1)
spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)

plt.xticks([0,15,30,45,60,76,90])
plt.xlabel('Minute',color='white',fontsize=16)
plt.ylabel('xG',color='white',fontsize=16)

ax.step(x=a_min,y=a_cumulative,color='#00933B',label=ateam,linewidth = 5, where= 'post')
ax.step(x=h_min,y=h_cumulative,color='#FFCD00',label=hteam,linewidth = 5, where= 'post')
```
