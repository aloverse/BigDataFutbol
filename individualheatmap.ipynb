```python
import numpy as np 
import pandas as pd
from mplsoccer import Pitch, Sbopen
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
parser = Sbopen()
df_match = parser.match(competition_id=43, season_id=3)
team = "Portugal"

match_id=df_match[df_match["home_team_name"] == team].iloc[0]['match_id']
```

```python
df_match[df_match["home_team_name"] == team].iloc[0]['away_team_name']
```

```python
df = parser.event(match_id)[0]
df.columns
```

```python
player = 'Cristiano Ronaldo dos Santos Aveiro'
mask_pass = df[(df['type_name'] == 'Pass')  & (df['team_name'] == team) & (df['player_name']==player)]
mask_pass
```

```python
mask_pass.columns
```

```python
mask_pass[['player_name','minute','second','x','y','outcome_name','end_x','end_y','pass_recipient_name']]
```

```python
#fig,ax =plt.subplots(figsize=(13.5,8))

#fig.set_facecolor('#22312b')
#ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb',pitch_color='#22312b',line_color='#c7d5cc')
fig,ax= pitch.draw(figsize=(16,11),constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

#pitch.draw(ax=ax)
plt.gca().invert_yaxis

kde=sns.kdeplot(
    x=mask_pass['x'],
    y=mask_pass['y'],
    fill= True,
    shade_lowest=False,
    alpha = .5,
    n_levels=10,
    cmap='magma'
)
```

```python
pitch = Pitch(pitch_type='statsbomb',pitch_color='#22312b',line_color='#c7d5cc')
fig,ax= pitch.draw(figsize=(16,11),constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

#pitch.draw(ax=ax)
plt.gca().invert_yaxis

kde=sns.kdeplot(
    x=mask_pass['x'],
    y=mask_pass['y'],
    fill= True,
    shade_lowest=False,
    alpha = .5,
    n_levels=10,
    cmap='magma'
)

for index,row in mask_pass.iterrows():
    if pd.isnull(row['outcome_name']):
        plt.plot((row['x'],row['end_x']),(row['y'],row['end_y']),color='green')
        plt.scatter(row['x'],row['y'],color='green')
    if pd.notnull(row['outcome_name']):
        plt.plot((row['x'],row['end_x']),(row['y'],row['end_y']),color='red')
        plt.scatter(row['x'],row['y'],color='red')

plt.xlim(0,120)
plt.ylim(0,80)

plt.title("Cristiano Roaldo Pass Map vs Morroco (Russia 2018)")
```
