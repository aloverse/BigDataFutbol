```python
import numpy as np 
import pandas as pd
from statsbombpy import sb
from mplsoccer import VerticalPitch
```

```python
events = sb.events(match_id=22912)
```

```python
events
```

```python
df =events[events["type"]=='Shot']
df
```

```python
df.columns
```

```python
df[['startx','starty']]= pd.DataFrame(df.location.tolist(),index=df.index)
df
```

```python
df['shot_outcome']
```

```python
liv='#c8102E'
liv_df = df[df["team"]=='Liverpool'].copy()


df_goals = liv_df[liv_df['shot_outcome'] == 'Goal'].copy()
df_saved = liv_df[liv_df['shot_outcome'] == 'Saved'].copy()
df_blocked = liv_df[liv_df['shot_outcome'] == 'Blocked'].copy()
df_off = liv_df[liv_df['shot_outcome'] == 'Off T'].copy()

pitch=VerticalPitch(
    half =True,
    pitch_color='grass',
    stripe=True,
    line_color='white'
)
fig , ax = pitch.draw(figsize=(10,8))
sc1 = pitch.scatter(df_goals['startx'],df_goals['starty'],
                    s = df_goals['shot_statsbomb_xg']*500,
                    ax=ax,
                    marker = 'football',
                    label="Goals")
sc2 = pitch.scatter(df_saved['startx'],df_saved['starty'],
                    s = df_saved['shot_statsbomb_xg']*500,
                    ax=ax,
                    c=liv,
                    marker = 'o',
                    label="Saved")
sc3 = pitch.scatter(df_blocked['startx'],df_blocked['starty'],
                    s = df_blocked['shot_statsbomb_xg']*500,
                    ax=ax,
                    c='yellow',
                    marker = 'h',
                    label="Blocked")
sc4 = pitch.scatter(df_off['startx'],df_off['starty'],
                    s = df_off['shot_statsbomb_xg']*500,
                    ax=ax,
                    c='grey',
                    marker = 'X',
                    label="Off")

ax.annotate("Shot Map Liverpool UCL Final 2018/2019",(40,65),fontsize=20,ha='center',c=liv)
ax.legend()
```
