```python
import numpy as np 
import pandas as pd
from statsbombpy import sb
from mplsoccer import VerticalPitch
```

```python
events = sb.events(match_id=22912)
```

```python
events
```

```python
df =events[events["type"]=='Shot']
df
```

```python
df.columns
```

```python
df[['startx','starty']]= pd.DataFrame(df.location.tolist(),index=df.index)
df
```

```python
liv='#c8102E'
liv_df = df[df["team"]=='Liverpool'].copy()
pitch=VerticalPitch(
    half =True,
    pitch_color='grass',
    stripe=True,
    line_color='white'
)
fig , ax = pitch.draw(figsize=(10,8))
sc = pitch.scatter(liv_df['startx'],liv_df['starty'],
                    s = liv_df['shot_statsbomb_xg']*500,
                    c=liv,
                    ax=ax,
                    label="Liverpool")

ax.annotate("Shot Map Liverpool UCL Final 2018/2019",(40,65),fontsize=20,ha='center',c=liv)
ax.legend()
```
